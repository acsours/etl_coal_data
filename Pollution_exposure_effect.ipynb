```python
import pandas as pd
```

### read files to dataframes and clean

```python
file_path_pm25="Resources/oecd_poll_exposure_mcgm3.csv"
file_path_mort="Resources/oecd_poll_effect_mortality.csv"
file_path_co2="Resources/oecd_air_ghg_CO2_t_pc.csv"
file_path_nox="Resources/oecd_air_ghg_NOx_kg_pc.csv"
file_path_sox="Resources/oecd_air_ghg_SOx_kg_pc.csv"
```

#### PM2.5

```python
poll_PM_df=pd.read_csv(file_path_pm25)
poll_PM_df.head()
```

```python
poll_PM_df.info()
```

```python
# drop column with Nan
poll_PM_df=poll_PM_df.drop(['Flag Codes'],axis=1)

poll_PM_df.head()
```

```python
# keep only 'LOCATION', 'TIME', and 'Value' columns
poll_PM_df=poll_PM_df[['LOCATION','TIME','Value']]
poll_PM_df.head()
```

#### Mortality

```python
poll_mortality_df=pd.read_csv(file_path_mort)
poll_mortality_df.head()
```

```python
poll_mortality_df.info()
```

```python
# keep only 'LOCATION', 'TIME', and 'Value' columns
poll_mortality_df=poll_mortality_df[['LOCATION','TIME','Value']]
poll_mortality_df.head()
```

#### CO2

```python
poll_CO2_df=pd.read_csv(file_path_co2)
poll_CO2_df.head()
```

```python
poll_CO2_df.info()
```

```python
# keep only 'LOCATION', 'TIME', and 'Value' columns
poll_CO2_df=poll_CO2_df[['LOCATION','TIME','Value']]
poll_CO2_df.head()
```

#### NOx and SOx

```python
poll_NOx_df=pd.read_csv(file_path_nox)
poll_NOx_df.head()
```

```python
poll_SOx_df=pd.read_csv(file_path_sox)
poll_SOx_df.head()
```

```python
poll_NOx_df.info()
```

```python
poll_SOx_df.info()
```

```python
# keep only 'LOCATION', 'TIME', and 'Value' columns
poll_NOx_df=poll_NOx_df[['LOCATION','TIME','Value']]
poll_NOx_df.head()
```

```python
# keep only 'LOCATION', 'TIME', and 'Value' columns
poll_SOx_df=poll_SOx_df[['LOCATION','TIME','Value']]
poll_SOx_df.head()
```

```python
# check for duplicates 
poll_PM_df[['LOCATION','TIME']].value_counts()
```

```python
# countries list
poll_PM_df['LOCATION'].unique()
```

```python
# need to drop groupped country 'G7M','WLD',G20','EU28','OECD','OECDE','EA19','EU27','EU27_2020'
country_drop_list=['G7M','WLD','G20','EU28','OECD','OECDE','EA19','EU27','EU27_2020']
```

```python
# "not in"
poll_PM_clean=poll_PM_df[~poll_PM_df['LOCATION'].isin(country_drop_list)]
poll_PM_clean.head()
```

```python
poll_mortality_clean=poll_mortality_df[~poll_mortality_df['LOCATION'].isin(country_drop_list)]

poll_CO2_clean=poll_CO2_df[~poll_CO2_df['LOCATION'].isin(country_drop_list)]

poll_NOx_clean=poll_NOx_df[~poll_NOx_df['LOCATION'].isin(country_drop_list)]

poll_SOx_clean=poll_SOx_df[~poll_SOx_df['LOCATION'].isin(country_drop_list)]
```

```python
# check available years
print(f"PM2.5 df {poll_PM_clean['TIME'].unique()}\n")
print(f"Mortality df {poll_mortality_clean['TIME'].unique()}\n")
print(f"CO2 df {poll_CO2_clean['TIME'].unique()}\n")
print(f"NOx df {poll_NOx_clean['TIME'].unique()}\n")
print(f"SOx df {poll_SOx_clean['TIME'].unique()}")
```

### Merge dataframes together

```python
# merge dataframes together
comb_df=pd.merge(poll_PM_clean,poll_mortality_clean, how='outer',on=['LOCATION','TIME'],suffixes=('_PM','_MR'))
comb_df
```

```python
comb_df.info()
```

```python
# check for duplicates
comb_df[['LOCATION','TIME']].value_counts()
```

```python
comb_df2=pd.merge(comb_df,poll_CO2_clean,how='left',on=['LOCATION','TIME'])
comb_df2
```

```python
# rename Value to Value_CO2
comb_df2=comb_df2.rename(columns={'Value':'Value_CO2'})
comb_df2.head()
```

```python
comb_df3=pd.merge(comb_df2,poll_NOx_clean,how='left',on=['LOCATION','TIME'])
comb_df3
```

```python
comb_full=pd.merge(comb_df3,poll_SOx_clean,how='left',on=['LOCATION','TIME'],suffixes=('_NOx','_SOx'))
comb_full
```

```python
# check for duplicates
comb_full[['LOCATION','TIME']].value_counts()
```

#### rename columns to match with others

```python
comb_full=comb_full.rename(columns={'LOCATION':'alpha_3', 'TIME':'year'})
comb_full
```

```python
# export combined dataframe to csv
comb_full.to_csv("output_data/combined.csv")
#comb_full[comb_full['alpha-3']=='USA'].to_csv("output_data/combined_usa.csv")
```
